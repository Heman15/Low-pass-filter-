```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
%matplotlib inline
```

```python
img = cv2.imread("C:\\Users\\saini\\Desktop\\testing\\Edge detection.jpg")
plt.imshow(img)  
```

```python
image = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)   #converting image to gray
plt.imshow(image,cmap = 'gray') 
```

```python
LPS = cv2.GaussianBlur(image,(5,5),0)
plt.imshow(LPS ,cmap = "gray")
```

```python
kernel = np.array([[-2 ,0,2],             #also called sobel filter
         [-1,0,1],
         [-2,0,2]])
kernel
```

```python
f ,(ax1,ax2) = plt.subplots(1,2, figsize = (20,10))

filtered_image = cv2.filter2D(image , -1,kernel)
ax1.set_title('original')
ax1.imshow(filtered_image,cmap = 'gray')

ax2.set_title('blur')
filtered_image_LPS = cv2.filter2D(LPS , -1,kernel)
ax2.imshow(filtered_image_LPS,cmap = 'gray')
```
